```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
p2l_merged_new = pd.read_feather("../data/2025-12-04-18:54:47_p2l_merged.feather")
p2l_merged_old = pd.read_feather("../data/p2l_merged.feather")
```

```python
l2p_merged_new = pd.read_feather("../data/2025-12-04-16:39:25_l2p_merged.feather")
l2p_merged_old = pd.read_feather("../data/l2p_merged.feather")
```

```python
p2l_merged_new.index = p2l_merged_new["sample"]
p2l_merged_new = p2l_merged_new.drop(columns=["sample"])

p2l_merged_old.index = p2l_merged_old["sample"]
p2l_merged_old = p2l_merged_old.drop(columns=["sample"])

l2p_merged_old.index = l2p_merged_old["sample"]
l2p_merged_old = l2p_merged_old.drop(columns=["sample"])

l2p_merged_new.index = l2p_merged_new["sample"]
l2p_merged_new = l2p_merged_new.drop(columns=["sample"])
```

```python
p2l_merged_new = p2l_merged_new[p2l_merged_new.columns[0:326457]]
p2l_merged_old = p2l_merged_old[p2l_merged_old.columns[0:326457]]
l2p_merged_old = l2p_merged_old[l2p_merged_old.columns[0:326457]]
l2p_merged_new = l2p_merged_new[l2p_merged_new.columns[0:326457]]
```

```python
analyze = ["GSX1_MIR1277"]

for edge in analyze:
    plt.scatter(l2p_merged_new[edge], p2l_merged_new[edge])
    plt.xlabel("L2P Edge Weight")
    plt.ylabel("P2L Edge Weight")
    plt.title(f"Per-Sample Edge Weight, L2P vs P2L, {edge}")
    plt.show()
```

```python
for edge in analyze:
    x_old, y_old = l2p_merged_old[edge].align(p2l_merged_old[edge])
    plt.scatter(x_old, y_old)
    plt.xlabel("L2P Edge Weight")
    plt.ylabel("P2L Edge Weight")
    plt.title(f"Per-Sample Edge Weight, L2P vs P2L, {edge}")
    plt.show()
```

```python
for edge in analyze:
    x_new, y_new = l2p_merged_new[edge].align(p2l_merged_new[edge])

    plt.scatter(x_old, y_old)
    plt.title(f"L2P vs P2L (OLD) {edge}")
    plt.show()

    x_new = l2p_merged_new[edge].sort_index()
    y_new = p2l_merged_new[edge].sort_index()

    plt.scatter(x_new, y_new)
    plt.title(f"L2P vs P2L (NEW) {edge}")
    plt.show()
```
