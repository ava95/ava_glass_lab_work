```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
motif = pd.read_csv('../data/motif.txt', sep="\t", header=None)
edges = pd.read_feather('../data/edges_diff_zscore.feather')

motif["TF"] = motif[0]
motif["Gene"] = motif[1]
motif["Score"] = motif[2]
motif = motif.drop(columns=[0,1,2])

tot_combs = len(motif['TF'].unique()) * len(motif['Gene'].unique())
real_combs = len(motif["TF"])
print(real_combs/tot_combs)
```

```python
real_edges = []

for idx, edge in edges.iterrows():
    tf = edge['TF']
    gene = edge['Gene']
    
    tf_df = motif[motif['TF'] == tf]
    
    if gene in list(tf_df['Gene']):
        real_edges.append(edge['edge'])
    
print(f"Number of total edges of interest: {edges.shape[0]}")
print(f"Number of edges of interest that actually exist in motif: {len(real_edges)}")
print(len(real_edges) / edges.shape[0])
```

```python
real_edges

real_edges_df = pd.DataFrame(real_edges)
real_edges_df = real_edges_df.rename(columns={0: "edge"})


real_edges_df[['TF', 'Gene']] = real_edges_df['edge'].str.split('_', expand=True)
print(real_edges_df.value_counts('Gene'))
```
